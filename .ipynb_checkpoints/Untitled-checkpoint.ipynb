```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_new_whole = image = cv2.imread(r'pic.jpg')

img_new_gray = cv2.cvtColor(img_new_whole, cv2.COLOR_BGR2GRAY)
image_new_template = cv2.imread(r'object_to_find.jpg')

s, m = image_new_template.shape[:-1]

result = cv2.matchTemplate(img_new_whole, image_new_template, cv2.TM_CCOEFF_NORMED)

threshold = 0.8

loc = np.where(result >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(img_new_whole, pt,(pt[0]+s, pt[1]+m),
    (0, 255,255), 2)
plt.imshow(img_new_whole)
```
